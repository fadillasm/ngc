H0 : The Popularity of Britney Spearsand Taylor Swift is statistically different 

H1 : The Popularity of Britney Spearsand Taylor Swift is statistically different 

the hypothesis testing:

```python
BS = df[df['Artist Name(s)']=='Britney Spears'][['Popularity']].groupby(['Popularity']).sum()
TS = df[df['Artist Name(s)']=='Taylor Swift'][['Popularity']].groupby(['Popularity']).sum()
```

```python
print('Popularity of Britney Spears =',df['Popularity'].mean())
print('Popularity of Taylor Swift =',df['Popularity'].mean())
```

```python
t_stat, p_val = stats.ttest_ind(df['Popularity'],df['Popularity'])
print('T-Statistic:',t_stat)
print('P-value:',p_val) #
```

```python
# standart deviation
std = df['Popularity'].std()
print('Standard Deviation of musics popularity:',std)
```

`Conclusion`: Based on p-value that is less than critical value, 0.05, so the H0 is rejected.
